Ejercicio 3: PANDIGITALES

Para el primer ejercicio, tomamos la lista de pandigitales de dígitos menores o iguales que 7, y seleccionamos aquellos que son primos:

```python
#función que calcula los pandigitales de k digitos:
def pandigitales(k):
    L=Permutations(k).list()
    pand=[]
    for J in L:
        J=[str(j) for j in J]
        pand+=[int(''.join(J))]
    return pand
```

```python
Pand=[]
for j in [1..7]:
    Pand+=pandigitales(j)
[j for j in Pand if Integer(j).is_prime()]
```

Último ejercicio: calculamos la suma de los que no son primos:

```python
Pand=[]
for j in [1..7]:
    Pand+=pandigitales(j)
sum([j for j in Pand if not Integer(j).is_prime()])
```
