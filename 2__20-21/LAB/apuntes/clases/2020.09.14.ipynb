Problema 1: El área de un polígono regular se puede calcular aplicando la fórmula Área = perímetro·apotema/2.
Escribe código de sage que calcule el perímetro y el área de un polígono regular utilizando los valores de dos variables: lados y longitud. ¿Es mayor un polígono regular de 2020 lados de 0.2020 centímetros de longitud, o un polígono regular de 20 lados de 20 centímetros de longitud?

```python
def func1(lados,longitud):
    apotema = longitud / (2*tan(pi/lados))
    perimetro = lados*longitud
    area = perimetro * apotema / 2
    return perimetro,area
```

```python
func1(4,2)
```

```python
p1,a1 = func1(2020,0.2020)
```

```python
p2,a2 = func1(20,20)
```

```python
print "Perímetro y área de un polígono regular de 2020 lados de 0.2020 centímetros de longitud: (", n(p1), ",", n(a1),")"
print "Perímetro y área de un polígono regular de 20 lados de 20 centímetros de longitud: (",n(p2),",",n(a2),")"
```

Es mayor el primero.

Problema 2: De un triángulo se conocen dos lados y el ángulo comprendido entre ambos. Asignando estos tres datos a sendasvariables, se pide escribir código que calcule:
i ) El semiperímetro del triángulo (sugerencia: utilizar el teorema del coseno).
ii ) El área del triángulo aplicando la fórmula de Herón.

```python
def func2(a,b,A):
    c = a**2 + b**2 - 2*a*b*cos(A)
    s = (a+b+c) / 2
    area = sqrt (s*(s-a)*(s-b)*(s-c))
    return s,area
```

```python
func2(1,1,pi/3)
```

Problema 3: Un cierto número de comensales van a repartirse, a partes iguales, una pizza redonda de cierto radio. Escribe el código necesario para calcular la superficie de pizza que le toca a cada persona en función del radio y de cuántas
personas son.

```python
def func3(radio, npersonas):
    
    return ((radio**2 * pi) / npersonas)
```

Problema 4: Tenemos una bicicleta con 10 marchas: cuando estamos pedaleando en la marcha n ∈ {1, . . . , 10}, cada pedalada corresponde a una fracción 1/n de vuelta de las ruedas (por ejemplo, en la marcha 2 se necesita dar 2 pedaladas para que las ruedas den una vuelta completa). Escribe código de sage que permita calcular cuántas pedaladas hay que dar, en función de la marcha que llevemos puesta, para recorrer una cierta distancia. Ten en cuenta que, obviamente, el radio de las ruedas de la bicicleta también es un factor a considerar.

```python
def func4(marcha, distancia, radio):
    distporpedalada = 2 * pi * radio / marcha
    return ceil(distancia/distporpedalada)
```

```python
func4(9,34.5,0.8)
```
