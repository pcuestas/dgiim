Problema 1: El rea de un polgono regular se puede calcular aplicando la frmula rea = permetroapotema/2.
Escribe cdigo de sage que calcule el permetro y el rea de un polgono regular utilizando los valores de dos variables: lados y longitud. Es mayor un polgono regular de 2020 lados de 0.2020 centmetros de longitud, o un polgono regular de 20 lados de 20 centmetros de longitud?

```python
def func1(lados,longitud):
    apotema = longitud / (2*tan(pi/lados))
    perimetro = lados*longitud
    area = perimetro * apotema / 2
    return perimetro,area
```

```python
func1(4,2)
```

```python
p1,a1 = func1(2020,0.2020)
```

```python
p2,a2 = func1(20,20)
```

```python
print "Permetro y rea de un polgono regular de 2020 lados de 0.2020 centmetros de longitud: (", n(p1), ",", n(a1),")"
print "Permetro y rea de un polgono regular de 20 lados de 20 centmetros de longitud: (",n(p2),",",n(a2),")"
```

Problema 2: De un tringulo se conocen dos lados y el ngulo comprendido entre ambos. Asignando estos tres datos a sendasvariables, se pide escribir cdigo que calcule:
i ) El semipermetro del tringulo (sugerencia: utilizar el teorema del coseno).
ii ) El rea del tringulo aplicando la frmula de Hern.

```python
def func2(a,b,A):
    c = a**2 + b**2 - 2*a*b*cos(A)
    s = (a+b+c) / 2
    area = sqrt (s*(s-a)*(s-b)*(s-c))
    return s,area
```

```python
func2(1,1,pi/3)
```

Problema 4: Tenemos una bicicleta con 10 marchas: cuando estamos pedaleando en la marcha n  {1, . . . , 10}, cada pedalada corresponde a una fraccin 1/n de vuelta de las ruedas (por ejemplo, en la marcha 2 se necesita dar 2 pedaladas para que las ruedas den una vuelta completa). Escribe cdigo de sage que permita calcular cuntas pedaladas hay que dar, en funcin de la marcha que llevemos puesta, para recorrer una cierta distancia. Ten en cuenta que, obviamente, el radio de las ruedas de la bicicleta tambin es un factor a considerar.

```python
def func4(marcha, distancia, radio):
    distporpedalada = 2 * pi * radio / marcha
    return ceil(distancia/distporpedalada)
```

```python
func4(9,34.5,0.8)
```
